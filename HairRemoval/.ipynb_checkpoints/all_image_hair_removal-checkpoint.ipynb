```python
def blackHighlights(image, edgeCoordinates):
    for c in edgeCoordinates:
        image[c[0], c[1]] = (0,0,0)
    
    return image
```

```python
def getCoordinates(index, radius, w, h):
#     startingCorner = [index[0] - radius, index[1] - radius]
    result = []
    for i in range(index[0] - radius, index[0] + radius + 1):
        if i >= h:
            break
        if i >= 0:
            for j in range(index[1] - radius, index[1] + radius + 1):
                if j >= w:
                    break
                if j >= 0:
                    result.append([i, j])

            
#     print(result)
    return result
```

```python
from statistics import mean 

def blurHairs(img2, coordinates, imageWithBlack):    
    
    h, w, c = img2.shape

    for c in coordinates:
        x = c[0]
        y = c[1]
        r = 0
        g = 0
        b = 0
        count = 0
        exceptions = 0
        tempCoordinates = getCoordinates([x,y], 10, w, h)#[[x, y+1], [x, y-1], [x+1, y], [x+1, y+1], [x+1, y-1], [x-1, y], [x-1, y+1], [x-1, y-1]]
        for c2 in tempCoordinates:
            x2 = c2[0]
            y2 = c2[1]
            try:
                if imageWithBlack[x2, y2][0] != 0 and imageWithBlack[x2, y2][1] != 0 and imageWithBlack[x2, y2][2] != 0:
                    r = r + img2[x2, y2][0]
                    g = g + img2[x2, y2][1]
                    b = b + img2[x2, y2][2]
                    count = count + 1
            except:
                exceptions = exceptions + 1
            if count > 0:
                img2[x,y] = [r/count, g/count, b/count]
    return img2
```

```python
from statistics import mean 

def findAverageColor(image, x, y, radius):
    count = 0
    r = 0
    g = 0
    b = 0
    for i in range(x - radius, y + radius + 1):
        if i >= h:
            break
        if i >= 0:
            for j in range(x - radius, y + radius + 1):
                if j >= w:
                    break
                if j >= 0:
                    n = image[i,j]
                    if(n[0] != 0 and n[1] != 0 and n[2] != 0):
#                         print(n)
#                         print(type(n))
                        r = r + n[0]
#                         print(r)
                        g = g + n[1]
                        b = b + n[2]
                        count = count + 1

        if count != 0:
            return [r/count, g/count, b/count]
        else:
            return np.array([0,0,0])
        

def blurHairs4(img, edges, imageWithBlack):
    
    indices = np.where(edges != [0])
    coordinates = zip(indices[0], indices[1])
    
    img2 = img
    h, w, c = img2.shape
    
    radius = 10
    
    prevX = 0
    prevY = 0
    
    print("avg", findAverageColor(imageWithBlack, prevX, prevY, 20))
    print("avg", type(findAverageColor(imageWithBlack, prevX, prevY, 20)))
    averageColor = findAverageColor(imageWithBlack, prevX, prevY, 15)
    
    countSinceLastUpdatedAverage = 0
    
    updateLimit = 10
    
    for c in coordinates:
        x = c[0]
        y = c[1]
        
        if (x != prevX and y != prevY) or countSinceLastUpdatedAverage >= updateLimit or np.count_nonzero(averageColor) > 0:
            averageColor = findAverageColor(imageWithBlack, x, y, 15)
            countSinceLastUpdatedAverage = 0
        else:
            countSinceLastUpdatedAverage = countSinceLastUpdatedAverage + 1
            prevX = x
            prevY = y
        
        if np.count_nonzero(averageColor) > 0:
            img2[x,y] = averageColor
        
        
    return img2
```
