```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
from sklearn import preprocessing
from sklearn.impute import MissingIndicator

continuous_data_columns = ["age_approx", "sun_exposure"]
discrete_data_columns = ["anatom_site_general", "diagnosis", "sex"]

dataset = pd.read_csv('C:/Users/ava/Desktop/KSU/Year 4/Spring 2020/Senior Project/new_skin_meta_aug2.csv')
# print(dataset.shape)
# dataset = dataset.dropna()
# print(dataset.shape)
```

```python
df = pd.DataFrame(dataset)
#    df.iloc[<rows>, <columns>]
discrete_df = df.iloc[:, 1:4]
continuous_df = df.iloc[:, 4:-1]

continuous_df = df[continuous_data_columns].values.astype(float)

min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
continuous_scaled = min_max_scaler.fit_transform(continuous_df)

# Run the normalizer on the dataframe
continuous_df = pd.DataFrame(continuous_scaled, columns=continuous_data_columns)

print(discrete_df)

col = 'sex'
for i in range(len(discrete_df[col])):
    if discrete_df[col][i] not in ['male', 'female']:
        discrete_df[col][i] = 'MISSING_GENDER'    

col = 'anatom_site_general'
for i in range(len(discrete_df[col])):
    if discrete_df[col][i] not in ['anterior torso', 'upper extremity', 'posterior torso', 'lower extremity', 'lateral torso', 'head/neck']:
        discrete_df[col][i] = 'MISSING_ANATOM_SITE_GENERAL'
        
col = 'diagnosis'
for i in range(len(discrete_df[col])):
    if discrete_df[col][i] not in ['nevus', 'dermatofibroma', 'basal cell carcinoma', 'vascular lesion', 'actinic keratosis', 'pigmented benign keratosis']:
#         print(discrete_df[col][i])
        discrete_df[col][i] = 'MISSING_DIAGNOSIS'
```
